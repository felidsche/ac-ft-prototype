```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
mpl.style.use("seaborn")
```

```python
tw_vals = [19.93, 16.23, 16.57, 17.62, 17.62]
tr_vals = [1.03, 1.08, 0.97, 1.07, 1.05]
tw_mean = np.mean(tw_vals)
tr_mean = np.mean(tr_vals)
exec_no = [1,2,3,4,5]
plt.figure(figsize=(8,6))
plt.xlabel("Execution number")
plt.xticks(exec_no, labels=exec_no)
plt.ylabel("Time (in minutes)")

plt.scatter(exec_no, tw_vals, linewidths=3)
plt.plot(exec_no, tw_vals, label="$T_w$")
plt.axhline(tw_mean, label =f"Avg $T_w$: {tw_mean}", linestyle="dotted", alpha=.7)

plt.scatter(exec_no, tr_vals, linewidths=3, marker="s")
plt.plot(exec_no, tr_vals, label="$T_r$")
plt.axhline(tr_mean, label =f"Avg $T_r$: {tr_mean}", linestyle="dashed", alpha=.7, color="green")

plt.legend()
plt.savefig("../../out/analysis/tw_tr/lineplot_measurement_exp_analytics30gb.pdf", format="pdf")
plt.savefig("../../out/analysis/tw_tr/lineplot_measurement_exp_analytics30gb.png", format="png")
```
